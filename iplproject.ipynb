# top 10 performance

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Top_100_batsman.xlsx")
top10 = df.sort_values(by='Runs',ascending=False).head(10)

plt.figure()
plt.xticks(rotation=45,ha='right')

colors = ['red', 'yellow', "#3b76ff", 'orange', 'skyblue','forestgreen','gold',"#7e1ea8","#f21e1e","#57057b"]

plt.bar(top10['PLAYER'],top10['Runs'],color=colors)
plt.xlabel('players')
plt.ylabel('Runs')
plt.title('top 10 performance')
plt.tight_layout()
plt.show()
```

# Strike rate vs batting avg

```python
import seaborn as sns

plt.figure()
sns.scatterplot(data=df,x='Avg',y='SR')
plt.title('Strike Rate vs Batting Average')
plt.xlabel('average')
plt.ylabel('strike rate')
plt.tight_layout()
plt.show()
```

# correlation heatmap of performance metrics

```python
numeric_df = df[['Mat','Runs','6s',100]]
plt.figure()
sns.heatmap(numeric_df.corr(),annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# Distribution of centuries

```python
sns.histplot(df[100],bins=50,kde=True)
plt.title('Distribution of centuries')
plt.xlabel('number of centuries')
plt.ylabel('number of players')
plt.show()
```

# Boundary contribution analysis

```python
df['boundary_runs'] = df['4s']*4 + df['6s']*6
df['boundary_%'] = (df['boundary_runs']/df['Runs'])*100
top5 = df.sort_values(by='Runs',ascending=False).head(5)
print(top5)
```
